```python
# Função para analisar a orientação de agendamento e classificar as informações de encaixe
def analisar_encaixe(orientacao):
    if pd.isna(orientacao):
        return "Informação Indisponível"
    orientacao = orientacao.lower()
    if 'encaixe' not in orientacao:
        return "Não Aceita Encaixe"
    if '1 encaixe' in orientacao:
        return "Aceita Somente 1 Encaixe"
    if '2 encaixes' in orientacao or 'dois encaixes' in orientacao:
        return "Aceita 2 Encaixes ou Mais"
    return "Aceita Encaixe"

# Aplicando a função para criar a nova coluna
df['Info_Encaixe'] = df['DS_ORIENTACAO_AGENDAMENTO'].apply(analisar_encaixe)

# Salvando o arquivo atualizado
updated_file_path = r'C:\Users\renan\Downloads\PROJETO_CLASSIFICAÇÃO\Prestadores_com_orientacao_de_agendamento.xlsx'
df.to_excel(updated_file_path, index=False)

updated_file_path
```
