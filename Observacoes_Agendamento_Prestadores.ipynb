```python
def analisar_encaixe(orientacao):
    if pd.isna(orientacao):
        return "Informação Indisponível"
    orientacao = orientacao.lower()
    if 'encaixe' not in orientacao:
        return "Não Aceita Encaixe"
    if '1 encaixe' in orientacao:
        return "Aceita Somente 1 Encaixe"
    if 'não encaixa' in orientacao:
        return "Não Aceita Encaixe"
    if '2 encaixes' in orientacao or 'dois encaixes' in orientacao or '2 encaixe' in orientacao:
        return "Aceita 2 Encaixes ou Mais"
    if '3 encaixes' in orientacao or 'tres encaixes' in orientacao or '3 encaixe' in orientacao:
        return "Aceita 3 Encaixes ou Mais"
    if '4 encaixes' in orientacao or 'quatro encaixes' in orientacao or '4 encaixe' in orientacao:
        return "Aceita 4 Encaixes ou Mais"
    if '5 encaixes' in orientacao or 'cinco encaixes' in orientacao or '5 encaixe' in orientacao:
        return "Aceita 5 Encaixes ou Mais"
    if '6 encaixes' in orientacao or 'seis encaixes' in orientacao or '6 encaixe' in orientacao:
        return "Aceita 6 Encaixes ou Mais"
    
    return "Aceita Encaixe"
```

```python
#aplicando a função analisar_encaixe criada para localizar a informação de encaixes na coluna de orientacao.
df ['Info_Encaixe'] = df ['DS_ORIENTACAO_AGENDAMENTO'].apply(analisar_encaixe)
```
